```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../sample.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("../landmarks/shape_predictor_68_face_landmarks.dat")
```

```python
result_image = image.copy()
faces = detector(result_image)
for face_index, face in enumerate(faces):
    landmark = predictor(result_image, face)
    for i in range(landmark.num_parts):
        x = landmark.part(i).x
        y = landmark.part(i).y
        
        cv2.putText(result_image, str(i), (x, y), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 0.3, (0, 255, 0))
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.show()
```
